```python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "SeaLLMs/SeaLLMs-v3-1.5B",
    load_in_4bit= False,
)
```

```python
from peft import load_peft_weights, set_peft_model_state_dict
```

```python
# load sft checkpoint
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
set_peft_model_state_dict(model, load_peft_weights('ppo_model'))
model= model.merge_and_unload()
```

```python
FastLanguageModel.for_inference(model)
```

```python
prompt= "### Question:\n{}\n### Answer:\n{}"
```

```python
def chat(text):
  text= prompt.format(text, "")
  inputs= tokenizer([text], return_tensors='pt').to("cuda")
  result= tokenizer.batch_decode(model.generate(**inputs, max_new_tokens= 256, do_sample= True, top_p=0.9, top_k= 100), skip_special_tokens= True)[0]


  return result.split('### Answer:')[1]
```

```python
print(chat("Đường cao tốc là đường gì?"))
```

```python
import gradio as gr
```

```python
def response(message, history):
  result= chat(message)

  return result
```

```python
mychatbot= gr.Chatbot(
    bubble_full_width= False,
    show_label= False,
    show_copy_button= True,
    height= 550,
    likeable=True,
)

demo = gr.ChatInterface(fn= response, chatbot= mychatbot, title= "Demo RLHF LawLLM", examples= ['Đường cao tốc là đường gì?'])

demo.launch(show_api= False, debug= True)
```
